```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Lawyers_Dataset.csv')

# Rename columns for clarity
data.columns = ['Web_Scraper_Order', 'Web_Scraper_Start_URL', 'Links', 'Links_Href', 'Name', 'Experience', 'Location', 'Languages', 'Practice_Areas', 'Courts', 'Pages', 'Pages_Href']

# Drop unnecessary columns
data = data.drop(['Web_Scraper_Order', 'Web_Scraper_Start_URL', 'Links', 'Links_Href', 'Pages', 'Pages_Href'], axis=1)

# Handle missing values
data['Experience'].fillna('Unknown', inplace=True)  # Replace missing experience with 'Unknown'
data['Courts'].fillna('Unknown', inplace=True)      # Replace missing courts with 'Unknown'

# Clean the 'Experience' column and convert it to numeric
data['Experience'] = data['Experience'].str.extract(r'(\d+)').fillna(0).astype(int)  # Extract years of experience as integers

# Split the 'Languages' column into a list of languages and handle NaN values
data['Languages'] = data['Languages'].apply(lambda x: [] if pd.isna(x) else x)

# If needed, perform further cleaning and preprocessing for the 'Practice_Areas' column

# Reset the index
data.reset_index(drop=True, inplace=True)

# Display the cleaned dataset
print(data.head())
```

In short, the case involves an individual who was appointed as an Officer in a bank and was also involved in a society formed to construct residential premises for bank employees. There were allegations of misappropriation of funds in the society, leading to inquiries and criminal charges against the individual. The bank suspended the individual, terminated his services, and later refused to reinstate him even after he was acquitted of the criminal charges. The individual filed a writ petition in the High Court, seeking to quash the bank's orders. Initially, the Single Judge granted relief, but the Division Bench of the High Court disagreed and dismissed the petition, citing delays and considering the merits of the case. This led to an appeal to the higher court.

```python
import pandas as pd

# Load the dataset and perform preprocessing as you did before
# (Load the dataset, drop unnecessary columns, handle missing values, etc.)

# Define a function to recommend lawyers based on user input
def recommend_lawyers_from_paragraph(input_paragraph, num_recommendations=5):
    # Here, you can use simple keyword matching to recommend lawyers
    # Extract keywords or phrases from the user input and match them with lawyer profiles
    keywords = ["civil", "criminal", "family", "divorce", "property", "litigation"]
    
    # Filter lawyers based on keyword matches
    matching_lawyers = data[data['Practice_Areas'].str.contains('|'.join(keywords), case=False, na=False)]
    
    # Sort lawyers by experience (you can use other criteria as well)
    recommended_lawyers = matching_lawyers.sort_values(by='Experience', ascending=False).head(num_recommendations)
    
    # Return the recommended lawyers as a list of dictionaries
    lawyer_details = recommended_lawyers.to_dict(orient='records')
    
    return lawyer_details

# Get user input as a paragraph
input_paragraph = input("Enter your legal case description paragraph: ")

# Get recommended lawyers based on the input paragraph
recommended_lawyers = recommend_lawyers_from_paragraph(input_paragraph)

# Display recommended lawyers
print("Recommended Lawyers:")
for idx, lawyer in enumerate(recommended_lawyers, start=1):
    print(f"Recommendation {idx}:")
    print(f"Name: {lawyer['Name']}")
    print(f"Experience: {lawyer['Experience']} years")
    print(f"Location: {lawyer['Location']}")
    print(f"Languages: {lawyer['Languages']}")
    print(f"Practice Areas: {lawyer['Practice_Areas']}")
    print()
```
