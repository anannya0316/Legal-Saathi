```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/LawyersData_withImages.csv')

# Rename columns for clarity
data.columns = ['Web_Scraper_Order', 'Web_Scraper_Start_URL', 'Links', 'Links_Href', 'Name', 'Experience', 'Location', 'Languages', 'Practice_Areas', 'Courts', 'Pages', 'Pages_Href','Profile_Pics_URL']

# Drop unnecessary columns
data = data.drop(['Web_Scraper_Order', 'Web_Scraper_Start_URL', 'Links', 'Links_Href', 'Pages', 'Pages_Href'], axis=1)

# Handle missing values
data['Experience'].fillna('Unknown', inplace=True)  # Replace missing experience with 'Unknown'
data['Courts'].fillna('Unknown', inplace=True)      # Replace missing courts with 'Unknown'

# Clean the 'Experience' column and convert it to numeric
data['Experience'] = data['Experience'].str.extract(r'(\d+)').fillna(0).astype(int)  # Extract years of experience as integers

# Split the 'Languages' column into a list of languages and handle NaN values
data['Languages'] = data['Languages'].apply(lambda x: [] if pd.isna(x) else x)

# If needed, perform further cleaning and preprocessing for the 'Practice_Areas' column

# Reset the index
data.reset_index(drop=True, inplace=True)

# Display the cleaned dataset
print(data.head())
```

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO

# Load the dataset and perform preprocessing as you did before
# (Load the dataset, drop unnecessary columns, handle missing values, etc.)

# Define a function to download and display images from URLs
def display_image_from_url(image_url):
    try:
        response = requests.get(image_url)
        img = Image.open(BytesIO(response.content))
        plt.imshow(img)
        plt.axis('off')  # Hide the axis
        plt.show()
    except Exception as e:
        print("Image not available")

# Define a function to extract numeric experience from the 'Experience' column
def extract_experience(experience_str):
    try:
        # Extract numeric part using regular expression
        numeric_part = re.search(r'\d+', experience_str)
        if numeric_part:
            return int(numeric_part.group())
        else:
            return 0  # Default to 0 if no numeric part found
    except Exception as e:
        return 0

# Define a function to recommend lawyers based on user input
def recommend_lawyers_from_paragraph(input_paragraph, num_recommendations=5):
    # Tokenize the user input paragraph into words
    input_tokens = re.findall(r'\b\w+\b', input_paragraph.lower())

    # Create an empty dictionary to store lawyer scores
    lawyer_scores = {}

    for idx, row in data.iterrows():
        # Tokenize the practice areas into words
        practice_areas_tokens = re.findall(r'\b\w+\b', row['Practice_Areas'].lower())

        # Calculate a score based on keyword matches
        score = sum(1 for token in input_tokens if token in practice_areas_tokens)

        # Add the lawyer and score to the dictionary
        lawyer_scores[row['Name']] = score

    # Sort lawyers by score in descending order
    sorted_lawyers = sorted(lawyer_scores.items(), key=lambda x: x[1], reverse=True)
    print("sorted_lawyers=>",sorted_lawyers)

    # Get the details of the recommended lawyers
    recommended_lawyers = [data[data['Name'] == lawyer[0]].to_dict(orient='records')[0] for lawyer in sorted_lawyers]

    # Sort recommended lawyers by experience (using the extracted numeric experience)
    recommended_lawyers.sort(key=lambda x: extract_experience(x['Experience']), reverse=True)

    return recommended_lawyers[:num_recommendations]

# Get user input as a paragraph
input_paragraph = input("Enter your legal case description paragraph: ")

# Get recommended lawyers based on the input paragraph
recommended_lawyers = recommend_lawyers_from_paragraph(input_paragraph)

# Display recommended lawyers with images from URLs
for idx, lawyer in enumerate(recommended_lawyers, start=1):
    print(f"Recommendation {idx}:")
    print(f"Name: {lawyer['Name']}")
    print(f"Experience: {lawyer['Experience']} years")
    print(f"Location: {lawyer['Location']}")
    print(f"Languages: {lawyer['Languages']}")
    print(f"Practice Areas: {lawyer['Practice_Areas']}")

    # Display the lawyer's image from the URL
    image_url = lawyer['Profile_Pics_URL']
    display_image_from_url(image_url)

    print()
```

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO
import spacy

# Load the spaCy language model
nlp = spacy.load("en_core_web_md")


# Load the dataset and perform preprocessing as you did before
# (Load the dataset, drop unnecessary columns, handle missing values, etc.)

# Define a function to download and display images from URLs
def display_image_from_url(image_url):
    try:
        response = requests.get(image_url)
        img = Image.open(BytesIO(response.content))
        plt.imshow(img)
        plt.axis('off')  # Hide the axis
        plt.show()
    except Exception as e:
        print("Image not available")

# Define a function to recommend lawyers based on user input
def recommend_lawyers_from_paragraph(input_paragraph, num_recommendations=5):
    # Tokenize and process the user input paragraph using spaCy
    input_doc = nlp(input_paragraph)

    # Create an empty dictionary to store lawyer scores
    lawyer_scores = {}

    for idx, row in data.iterrows():
        # Tokenize and process the lawyer's practice areas using spaCy
        practice_area_doc = nlp(row['Practice_Areas'].lower())

        # Calculate a relevance score based on keyword matches
        relevance_score = input_doc.similarity(practice_area_doc)
        lawyer_scores[row['Name']] = relevance_score

    # Sort lawyers by relevance score in descending order
    sorted_lawyers = sorted(lawyer_scores.items(), key=lambda x: x[1], reverse=True)[:num_recommendations]
    print("Lawyers scores",sorted_lawyers)

    # Get the details of the recommended lawyers
    recommended_lawyers = [data[data['Name'] == lawyer[0]].to_dict(orient='records')[0] for lawyer in sorted_lawyers]

    return recommended_lawyers

# Get user input as a paragraph
input_paragraph = input("Enter your legal case description paragraph: ")

# Get recommended lawyers based on the input paragraph
recommended_lawyers = recommend_lawyers_from_paragraph(input_paragraph)

# Display recommended lawyers with images from URLs
for idx, lawyer in enumerate(recommended_lawyers, start=1):
    print(f"Recommendation {idx}:")
    print(f"Name: {lawyer['Name']}")
    print(f"Experience: {lawyer['Experience']} years")
    print(f"Location: {lawyer['Location']}")
    print(f"Languages: {lawyer['Languages']}")
    print(f"Practice Areas: {lawyer['Practice_Areas']}")

    # Display the lawyer's image from the URL
    image_url = lawyer['Profile_Pics_URL']
    display_image_from_url(image_url)

    print()
```
